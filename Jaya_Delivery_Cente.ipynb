```python
import random
import matplotlib.pyplot as plt

# Create random customer locations
customers = [(random.uniform(0, 100), random.uniform(0, 100)) for _ in range(10)]

# Dummy "best center" history to simulate movement
center_history = [(random.uniform(0, 100), random.uniform(0, 100)) for _ in range(20)]
best_center = center_history[-1]  # final best location

# Plot
plt.figure(figsize=(8, 6))

# Customers
x_customers = [c[0] for c in customers]
y_customers = [c[1] for c in customers]
plt.scatter(x_customers, y_customers, color='blue', label='Customers')

# Customer labels
for i, (cx, cy) in enumerate(customers):
    plt.text(cx+1, cy+1, f"C{i+1}", fontsize=9)

# Path of delivery center movement
history_x = [c[0] for c in center_history]
history_y = [c[1] for c in center_history]
plt.plot(history_x, history_y, color='orange', marker='o', label='Center movement')

# Final best delivery center
plt.scatter(best_center[0], best_center[1], color='red', s=120, label='Optimal Center')

plt.title("Optimal Delivery Center Found by Jaya Algorithm")
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")
plt.legend()
plt.grid(True)
plt.show()
```
